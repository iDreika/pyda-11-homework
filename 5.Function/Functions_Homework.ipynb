Вы работаете секретарем и вам постоянно приходят различные документы. Вы должны быть очень внимательны, чтобы не потерять ни один документ. Каталог документов хранится в следующем виде:

```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
```

Перечень полок, на которых находятся документы хранится в следующем виде:

```python
directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}
```

Необходимо реализовать пользовательские команды (внимание! это не название функций, которые должны быть выразительными, а команды, которые вводит пользователь, чтобы получить необходимый результат):
  
- p – people – команда, которая спросит номер документа и выведет имя человека, которому он принадлежит;  
- s – shelf – команда, которая спросит номер документа и выведет номер полки, на которой он находится  
*Правильно обработайте ситуации, когда пользователь будет вводить несуществующий документ*.
- l – list – команда, которая выведет список всех документов в формате   
**passport "2207 876234" "Василий Гупкин"**  
- as – add shelf – команда, которая спросит номер новой полки и добавит ее в перечень. *Предусмотрите случай, когда пользователь добавляет полку, которая уже существует*.
- d – delete – команда, которая спросит номер документа и удалит его из каталога и из перечня полок. *Предусмотрите сценарий, когда пользователь вводит несуществующий документ*;  
- m – move – команда, которая спросит номер документа и целевую полку и переместит его с текущей полки на целевую. *Корректно обработайте кейсы, когда пользователь пытается переместить несуществующий документ или переместить документ на несуществующую полку*;  
- a – add – команда, которая добавит новый документ в каталог и в перечень полок, спросив его номер, тип, имя владельца и номер полки, на котором он будет храниться. *Корректно обработайте ситуацию, когда пользователь будет пытаться добавить документ на несуществующую полку*.

```python
print('\033[1mДобро пожаловать в \033[94mCity17\033[0m!')
        
# Объявление функций
def people(doc=documents):
    """
    Функция, которая спросит номер документа и выведет имя человека, которому он принадлежит
    """
    p = input('\nЗапрос имени. \nВведите номер документа: ')
    i = 0
    for element in doc:
        if element['number'] == p:
            print('\nИскомое имя: \033[1m', element['name'], '\033[0m')
        else:
            i += 1
        if len(doc) == i:
            print('\n\033[1m\033[91mУказанный документ не найден!\033[0m\n')

def shelf(directory):
    """
    Функция, которая спросит номер документа и выведет номер полки, на которой он находится
    """
    p = input('\nЗапрос номера полки. \nВведите номер документа: ')
    i = 0
    for element in directory:
        for number in directory[element]:            
            if number == p:
                print('\nНомер полки:', element)
        i+=1
    if len(directory) == i:
        print('\n\033[1m\033[91mУказанный документ не найден!\033[0m\n')
            
def list_f(doc=documents):
    """
    Функция, которая выведет список всех документов в формате
    """
    print('\n')
    for element in doc:
        print('\033[1m', element['type'], ' "', '\033[94m', element['number'], '\033[0m\033[1m', '" ', element['name'], '\033[0m',sep='')
        
def add_shelf(directory=directories):
    """
    Функция, которая спросит номер новой полки и добавит ее в перечень
    """
    p = input('\nСоздание полки. \nВведите номер полки: ')
    if p in directory.keys():
        print('\n\033[1m\033[91mУказанная полка уже существует!\033[0m\n')
    else:
        directory[p] = [] 
        print('\nПолка номер ', '\033[1m', p, '\033[0m', ' создана', sep='')

def delete(doc=documents, directory=directories):
    """
    Функция, которая спросит номер документа и удалит его из каталога и из перечня полок
    """
    p = input('\nЗапрос на удаление. \nВведите номер документа для удаления: ')
    i = 0
    for element in doc:
        if element['number'] == p:
                documents.remove(element)
                for shelf in directory.values():
                    if p in shelf:
                        shelf.remove(p)
                        print('\nДокумент номер ', '\033[1m', p, '\033[0m', ' удалён', sep='')
        i+=1
    if len(documents) == i:
        print('\n\033[1m\033[91mУказанный документ не найден!\033[0m\n')

def move(directory=directories):
    """
    Функция, команда, которая спросит номер документа и целевую полку и переместит его с текущей полки на целевую
    """        
    i = 0
    f_number = input('\nЗапрос на перемещение. \nВведите номер документа для перемещения: ')    
    f_shelf = input('Введите номер конечной полки: ')
    if f_shelf not in directory:
        print('\n\033[1m\033[91mУказанная полка не существует!\033[0m\n')
    else:
        for element in directory:
            if f_number in directory[element]:
                directories[element].remove(f_number)
                directories[f_shelf].append(f_number)
                print('\nДокумент номер ', '\033[1m\033[94m', f_number, '\033[0m', ' добавлен на полку номер ', '\033[1m\033[94m', f_shelf, '\033[0m', sep='')
                break
            i += 1
    if len(directory) == i:
        print('\n\033[1m\033[91mУказанный документ не найден!\033[0m\n')
            
def add(doc=documents, directory=directories):
    """
    Функция, которая добавит новый документ в каталог и в перечень полок, спросив его номер, тип, имя владельца и номер полки, на котором он будет храниться
    """   
    i = 0
    f_number = input('\nЗапрос на добавление документа. \nВведите номер документа для добавления: ')
    f_type = input('Введите тип документа: ')
    f_name = input('Введите имя владельца документа: ')
    f_shelf = input('Введите номер полки для хранения: ')
    if f_shelf not in directory:
        print('\n\033[1m\033[91mУказанная полка не существует!\033[0m\n')
    else:
        documents.append({'type': f_type, 'number': f_number, 'name': f_name})
        directories[f_shelf].append(f_number)
        print('\nДокумент \033[1m', f_type, ' ""\033[94m',f_number, '"" \033[0m\033[1m', f_name,' \033[0mсоздан', sep='')
        print('Документ номер ', '\033[1m\033[94m', f_number, '\033[0m', ' добавлен на полку номер ', '\033[1m\033[94m', f_shelf, '\033[0m', sep='')
       
                
    
        
# Цикл исполнения запросов            
while 3>2:
    command = input('\n\033[95mВведите команду: \033[0m')
    if command == 'p':
        people()
    elif command == 's':   
        shelf()            
    elif command == 'l':   
        list_f()            
    elif command == 'as':   
        add_shelf()             
    elif command == 'd':   
        delete()
    elif command == 'm':   
        move() 
    elif command == 'a':   
        add()
    else:
        print('\nТакой команды не существует!')
        continue
    variant = input('\nЖелаете продолжить? (N/Y): ')
    if variant == 'Y' or variant == 'y':
        continue
    elif variant == 'N' or variant == 'n':
        print('\n\033[94m\033[1mВсего наилучшего!')
        break
    else:
        print('\nДанный ответ не предусмотрен! Понимаем как "\033[1m\033[91mНет\033[0m".\n\033[94m\033[1mВсего наилучшего!')
        break
```
