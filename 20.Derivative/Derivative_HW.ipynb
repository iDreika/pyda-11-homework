### Уровень 0

Посчитайте значение 1-ой и 2-ой производной функции:

f(x) = x^5 + 4sin(2x) + cos(3x+3) в точке x = 1



```python
from scipy.misc import derivative
import numpy as np

def Inv_f(x):
    return x**5 + 4*np.sin(2*x) + np.cos(3*x+3)
    

print('Зачение 1-ой производной функции f(x) в точке х=1: ', derivative(Inv_f, 1.0, dx=1e-6))
print('Зачение 2-ой производной функции f(x) в точке х=1: ',derivative(Inv_f, 1.0, dx=1e-6, n=2))
```
