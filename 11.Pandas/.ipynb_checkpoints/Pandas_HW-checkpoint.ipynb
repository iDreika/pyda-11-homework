### Задание 1
Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.



```python
import numpy as np
import pandas as pd

movies = pd.read_csv('movies.csv')
ratings = pd.read_csv('ratings.csv')
ratings.head()

joined = ratings.merge(movies, on='movieId', how='left')
joined.query(f'rating == {5.0}').groupby('title').count().sort_values('rating', ascending=False).head(1)
```

### Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.



```python
import sys

power = pd.read_csv('power.csv')

sum = power[ (power['country'] == 'Lithuania') | 
            (power['country'] == 'Latvia') | 
            (power['country'] == 'Estonia')][(power['category'] == 4 ) | 
                                            (power['category'] == 12) | 
                                            (power['category'] == 21  )][power['quantity'] >= 0].sum()
print('Суммарное потребление стран Прибалтики = ', sum['category'])
```
