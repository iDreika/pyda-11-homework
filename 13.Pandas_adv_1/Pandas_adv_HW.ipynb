### Задание 1

Для датафрейма log из материалов занятия создайте столбец source_type по следующим правилам:

<ul>
<li>если источник traffic_source равен yandex или google, то в source_type ставится organic</li>
<li>для источников paid и email из России - ставим ad</li>
<li>для источников paid и email не из России - ставим other</li>
<li>все остальные варианты берем из traffic_source без изменений</li>
</ul>

```python
import pandas as pd

def source_t_f(row):
    """
    Функция, определяющая source_type.
    Входными данными является DataFrame - Row.
    """
    source_t_v = row['traffic_source']
    if row['traffic_source'] == 'yandex' or row['traffic_source'] == 'google':
        source_t_v = 'organic'
    elif row['traffic_source'] == 'paid' or row['traffic_source'] == 'email':
        if row['region'] == 'Russia':
            source_t_v = 'ad'  
        else:
            source_t_v = 'other'
    return source_t_v
    
log = pd.read_csv('visit_log.csv', sep=';')
log['source_type'] = log.apply(source_t_f, axis=1)

log.loc[log.traffic_source == 'email']
```

### Задание 3

В датафрейме data создайте столбец lemmas, в котором вычислите леммы поисковых запросов из столбца keyword. Леммы должны иметь строковый тип.

```python
from pymystem3 import Mystem

def lemm_f (key):
    m = Mystem()
    lemmas = m.lemmatize(key)
    final_l = set(lemmas)
    final_l.remove('\n')
    final_l.remove(' ')
    lemmas = list(final_l)
    final_l = ''
    for word in lemmas:
        final_l += word + ','     
    return final_l

data = pd.DataFrame({
    'keyword': ['курс гривны к рублю', 'доллары в рубли', '100 долларов в рублях', 'курс рубля'],
    'shows': [125076, 114173, 97534, 53546],
})

data['lemmas'] = data['keyword'].apply(lemm_f)
data.head()
```
