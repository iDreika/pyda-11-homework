```python
import pandas as pd
import re
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
from pylab import rcParams
rcParams['figure.figsize'] = 15,8

def date_conv_y(date_v):
    try:
        out_y = datetime.strptime(str(date_v).strip(), '%B %d, %Y').strftime('%Y')
    except:
        out_y = None
    return out_y

netflix = pd.read_csv('netflix_titles.csv')
netflix['year_added'] = netflix['date_added'].apply(date_conv)
netflix.head()
```

```python
movie_df = netflix.loc[netflix['type'] == 'Movie']
movie_df['duration_m'] = movie_df['duration'].apply(lambda duration: int(re.findall(r'\d+',duration)[0]))

listed_in = movie_df.listed_in.value_counts().sort_values(ascending = False).head(5).index.values
listed_in
ax = sns.boxplot(x='listed_in', y='duration_m', 
            data=movie_df[movie_df.listed_in.isin(listed_in)])
```

Данные ящики позволяют сделать вывод о трендовой продолжительности различных жанров кино. Однако в частности интересно что пределом в продолжительности вне зависимости от жанра около 190 минут.

```python
tv_df = netflix.loc[netflix['type'] == 'TV Show']
tv_df.groupby('duration').count().plot(kind='pie', y='show_id')
```

Пирог позволяет четко видеть, что по мере увеличения количества сезонов уменьшается его шанс на дальнейшую жизнь. Другими словами создателям достаточно сложно создавать долгоживущие проекты.

```python
tv_df['seas_dur'] = tv_df['duration'].apply(lambda duration: int(re.findall(r'\d+',duration)[0]))

countries = tv_df.country.value_counts().sort_values(ascending = False).head(5).index.values
ax = sns.boxplot(x='country', y='seas_dur', 
            data=tv_df[tv_df.country.isin(countries)])
```

Под данным ящикам видно, что преимыщественно все страны производят односезонные сериалы. Также важно отметить что в США, Великобритании и Японии имеют место многосезонные шоу.

```python
per_year = (netflix
         .sort_values('year_added', ascending = False)
         .groupby(['year_added','type'])
         .count()[['show_id']]
           )
per_year.unstack().plot(kind='bar')
```

Данный график помогает сделать вывод о том что Netflix из года в год увеличивает размер наращиваемого контента. Так же важно отметить то, преимущественно в сервис добавляются фильмы - вероятно протсо потому что фильмов де-факто больше чем сериалов или TV шоу.
